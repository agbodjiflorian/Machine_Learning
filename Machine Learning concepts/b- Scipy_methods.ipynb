```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

# Scipy Interpolate

# 1. Interpolation

```python
x = np.linspace(0, 10, 10)
y = x**2
plt.scatter(x, y);
```

```python
from scipy.interpolate import interp1d
```

```python
# interp1d nous retourne une fonction f(...)
f = interp1d(x, y, kind = 'linear')

new_x = np.linspace(0, 10, 30)
result = f(new_x)

plt.scatter(x, y)
plt.scatter(new_x, result, c = 'r')
```

# 2. Optimization

```python
# curve_fit: il utilise la méthode des moindres carrés pour trouver les bons paramètres d'un modèle f

x = np.linspace(0, 2, 100)
y = 1/3*x**3 - 3/5 * x**2 + 2 + np.random.randn(x.shape[0])/20

plt.scatter(x, y);
```

```python
# f(x) = ax**3 + bx**2 + cx + d
def f(x, a, b, c, d):
    return ((a * x**3) + (b * x**2) + (c * x) + d)
```

```python
from scipy import optimize
```

```python
params, param_cov = optimize.curve_fit(f, x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]), c = 'g', lw = 3)
```

# 3. Minimisation

```python
def f(x):
    return ((x**2) + 15*np.sin(x))
```

```python
x = np.linspace(-10, 10, 100)

plt.plot(x, f(x));
```

```python
# exécute un algorithme de minimisation selon un point de départ x0
x0 = -5
result = optimize.minimize(f, x0 = x0).x
```

```python
plt.plot(x, f(x), lw = 3, zorder = -1)
plt.scatter(result, f(result), s = 100, c='r', zorder = 1)
plt.scatter(x0, f(x0), s=200, marker='+', c='g', zorder = 1)
plt.show()
```
